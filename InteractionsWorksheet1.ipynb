### Experiment Detail:
### A farmer conducted an experiment. He fed some of his chickens with normal feed, some with a special feed X and some with a special feed Y over a period of 6 months.
### The histogram below show the outcomes of the experiment.


```python
# The histogram:
# Import matplotlib.pyplot
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

# data to plot
n_groups = 6
y1=(500,750,1000,1500,2000,2250)
y2=(500,600,800,1200,1800,2250)
y3=(500,800,1150,1900,2150,2500)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.8

rects1 = plt.bar(index, y1, bar_width, alpha=opacity, color='r', label='Special Feed X')

rects2 = plt.bar(index + bar_width, y2, bar_width, alpha=opacity, color='g',label='Normal Feed')

rects3 = plt.bar(index + 2*bar_width, y3, bar_width, alpha=opacity, color='b',label='Special Feed Y')

plt.xlabel('Month')
plt.ylabel('Chicken Weight in Gram')
plt.title('Chicken Growth with Different Feed')
plt.xticks(index + bar_width, ('1', '2', '3', '4','5','6'))
plt.legend()

plt.tight_layout()
plt.show()
```

### 2) There are three groups of variables in an experiment namely dependant variable, independant variable and control variable.
### 3) Depedent Variable is typcially the y variable in the graph.
### 4) Indepedant Variable is typically the x variable in the graph.
### 5) The Dependant Variable changes in response to changes in the Independant Variable.
