### Experiment Detail:
### Anne Wee conducted an experiment to investigate how the temperature of water influences the photosynthesis rate of an aquatic plant.
### She set up the experiment as above. She varied the temperature of pond water at 5 degree C interval from 20 degree C to 40 degree C. She recorded the volume of oxygen collected in the test tube after 2 hours and plotted a graph to show the effect of the temperature of pond water on the photosynthesis rate. The graph is shown below.

```python
# Click 'Run' button to show the graph..
# Import matplotlib.pyplot
%matplotlib notebook
import matplotlib.pyplot as plt

plt.ion()

x1=(16,18,20,22,24,25,26)
y1=(10,12,15,19,23,24.5,25)
x2=(26,27,28,30,32,34,36,38,40,42,44,46)
y2=(25,24.5,23.5,22,19.5,17.5,15,13,11.5,11,10,10)
# Plot in blue the % of degrees awarded to women in the Physical Sciences
plt.plot(x1,y1, color='black')

# Plot in red the % of degrees awarded to women in Computer Science
plt.plot(x2, y2, color='black')

# Display the plot
plt.title('Photosynthesis Rate Vs Temperature')
plt.xlabel('Temperature of Water(Degree Celsius)')
plt.ylabel('Amount of Oxygen Collected(mm)')
plt.axis((10,50,0,35))

plt.tight_layout()
plt.show()
```
