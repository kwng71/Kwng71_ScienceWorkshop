### Main Learning Points:
### 1) Ensure you use the following key words when you answer the questions:
#### -transparent
#### -most light
#### -some light
#### -no light
#### -light intensity
### 2) Three optical properties of materials are transparent, transluscent and opaque.
### 3) Transparent materials allow most light to pass through them.
### 4) Transluscent materials allow some light to pass through them in a diffused manner.
### 5) Opaque materials allow no light to pass through them.
### 6) Amount of light measured is the Dependant Variable
### 7) Type of material is the Independant Variable.
