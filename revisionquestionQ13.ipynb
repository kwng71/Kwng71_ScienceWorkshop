| Time(Minutes) | Temperature of air inside the small pot(Degree C)|

```python
# Graph
%matplotlib notebook
import matplotlib.pyplot as plt

plt.ion()

WaterEvaporated=(10,30,60,90)
AirTemperature=(30,28,27,25)


# Plot in red the % of degrees awarded to women in the Physical Sciences
plt.plot(WaterEvaporated,AirTemperature, color='red', label='Air Temperature')


# Display the plot
plt.title('Rate of Evaporative Cooling')
plt.xlabel('Amount of Water Evaporates(mg)')
plt.ylabel('Air Temperature in Degree C')
plt.axis((0,100,20,35))
plt.legend()

plt.tight_layout()
plt.show()
```

### Main Learning Points:
### 1) Ensure you use the following key words when you answer the questions:
#### -Heat
#### -liquid form

### 2) Heat in the air could be removed by evaporation of water. 
### 3) During evaporation, water absorbs energy and changes from liquid form to vapour form.
### 4) Tempurature can be measured with:
#### -thermometer
#### -temperature sensors with datalogger
### 5) Cooling effect(Air temperature) is the Dependant Variable.
### 6) Evaporation rate(Mass of water evaporated in a period) is the Independant Variable.
### 7) The larger the amount of water evaporated, the larger the amount of heat removed, the lower the air temperature.
